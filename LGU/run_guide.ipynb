# 🎬 영화 리뷰 데이터 수집 및 분석 파이프라인 (터미널 복사용)

이 노트북 파일은 **노트북 내에서 직접 코드를 실행하는 용도가 아닙니다.**
아래 각 코드 셀에 있는 **명령어를 마우스로 복사하여, 실제 터미널이나 아나콘다 프롬프트(Anaconda Prompt) 창에 붙여넣고 실행**하기 위한 가이드 문서입니다.

### 1. 작업 디렉토리 이동 및 가상환경 활성화
아나콘다 프롬프트를 열면 기본 위치가 다르기 때문에, `LGU` 폴더로 이동해야 합니다.

✨ **경로를 자동으로 입력하는 가장 쉬운 방법 (드래그 앤 드롭)**
1. 아나콘다 프롬프트에 `cd ` (cd 뒤에 스페이스바 한 칸 띄우기) 까지만 입력합니다.
2. 윈도우 파일 탐색기에서 `LGU` 폴더를 마우스로 꾹 눌러서 아나콘다 프롬프트 검은 창 위로 끌어다 놓습니다(드래그 앤 드롭).
3. 경로가 자동으로 텍스트로 입력되면 엔터(Enter)를 칩니다.
4. 경로 이동 후 `conda activate PJ` 를 입력하여 가상환경을 켭니다.

```bash
# (예시) 터미널 화면:
# cd C:\Users\...\TeamProject\movie_reviews\LGU
conda activate PJ
```

### 2. 리뷰 데이터 스크래핑
Watchapedia에서 리뷰 데이터를 수집하여 `data/original` 폴더에 저장합니다.

```bash
python scraper.py
```

### 3. 데이터 정제 및 중복 제거
수집된 원본 데이터의 중복을 제거하고 정제하여 `data/cleaned_total_reviews.json`을 생성합니다.

```bash
python cleaner.py
```

### 4. 데이터 무결성 검사
데이터 내 빈 값이 있는지 확인하고, 문제가 있다면 `data_error_report.txt`를 생성합니다.

```bash
python checker.py
```

### 5. 누락 데이터 복구 시도 (API & 수동)
`checker.py`에서 발견된 오류를 수정하기 위해 `fixer.py`를 실행합니다.
이 스크립트는 **먼저 TMDB와 KOBIS API를 통해 누락된 데이터를 자동으로 검색하여 채워 넣습니다.**
만약 API로도 찾지 못한 데이터가 있다면 `manual_fix_form.txt` 파일을 생성한 뒤 **해당 텍스트 파일을 자동으로 엽니다.**
화면에 뜬 텍스트 파일의 빈 칸을 채우고 저장(`Ctrl+S`)하신 뒤, 터미널 창으로 돌아와서 엔터(Enter) 키를 누르면 마저 반영됩니다.

```bash
python fixer.py
```

⚠️ **주의**: 아래 분석 작업을 실행하기 전에 반드시 **Ollama 서버가 켜져 있어야** 합니다.
(아나콘다 프롬프트 창을 하나 더 열고 `ollama serve`를 입력해두시면 됩니다.)

### 6. 로컬 LLM을 이용한 리뷰 분석 및 태깅
Gemma4 모델을 사용하여 리뷰를 분석하고 `data/tagged_reviews.json`을 생성합니다.

```bash
python classifier.py
```

### 7. 룰베이스(Fallback)로 임시 태깅된 리뷰들 재분석
이전 단계에서 에러로 인해 룰베이스로 임시 태깅된 항목만 골라서 다시 LLM 분석을 시도합니다.

```bash
python re_classifier.py
```
